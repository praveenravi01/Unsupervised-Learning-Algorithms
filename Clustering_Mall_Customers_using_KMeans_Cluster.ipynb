```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ds=pd.read_csv("Mall_Customers.csv")
ds
```

```python
# dt=ds.drop("CustomerID",axis=1)
# dt=dt.drop("Genre",axis=1)
# dt=dt.drop("Age",axis=1)
ds.isnull().sum()
```

```python
x=ds[["Annual_Income_(k$)","Spending_Score"]].values
```

```python
x[:5]
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(10,5))
plt.plot(range(1,11),wcss)
plt.title("the Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("wcss")
plt.show
```

```python
sns.boxplot(ds)
```

```python
kmeans=KMeans(n_clusters=5,init="k-means++",random_state=40)
y_kmeans=kmeans.fit_predict(x)
```

```python
y_kmeans
```

```python
kmeans.inertia_
```

```python
kmeans.labels_
```

```python
plt.figure(figsize=(15,7))
plt.scatter(x[y_kmeans==0,0], x[y_kmeans==0,1], color="yellow", label="cluster1", s=50)
plt.scatter(x[y_kmeans==1,0], x[y_kmeans==1,1], color="grey", label="cluster2", s=50)
plt.scatter(x[y_kmeans==2,0], x[y_kmeans==2,1], color="blue", label="cluster3", s=50)
plt.scatter(x[y_kmeans==3,0], x[y_kmeans==3,1], color="green", label="cluster4", s=50)
plt.scatter(x[y_kmeans==4,0], x[y_kmeans==4,1], color="magenta", label="cluster5", s=50)

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color="red", label="Centroids", s=150, marker=",")
plt.grid(False)
plt.title("Cluster of Customers")
plt.xlabel("CustomerID")
plt.ylabel("Spending_Score")
plt.legend()
plt.show()

```
